# get the value of a selected parameter as a variable  

```python
Height=Your_Data["Height"].values
Weight=Your_Data["Weight"].values
```

# plot Weight vs. Height using  '.'

```python
plt.plot(Height,Weight, '.')
```

# make the plot more informative by adding labels, title ,  And save it with a name 

```python
plt.plot(Height,Weight, '.')
plt.title('Correlation', fontsize=19)
plt.xlabel('Height')
plt.ylabel('Weight')

fig_name='fig-correlation'  # choose a name for you plot to save '
plt.savefig(fig_name+'.png', format='png', dpi=100) 
```
