```python
# Read file 'Height_Weight_NaN.csv' as HW_data_NaN
HW_data_NaN= pd.read_csv('Height_Weight_NaN.csv')
print('the file loaded')
```

```python
# see the table
HW_data_NaN
```

```python
# see information of the file
HW_data_NaN.info()
```

```python
# a new name for the corrected file
#use  'ffill' if you want to propagate the last valid observation forward 

HW_data_NaN_fillmed= HW_data_NaN.fillna(HW_data_NaN.median())
```

```python
HW_data_NaN_fillmed.to_csv('HW_data_NaN_fillmed.csv',index=False)
print ('The new file was saved')
```

```python
# a new name for the corrected file
HW_data_NaN_drop= HW_data_NaN.dropna()
```
